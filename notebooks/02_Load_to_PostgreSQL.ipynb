```python
import pandas as pd
from sqlalchemy import create_engine

# Ruta al archivo limpio
DATA_PATH = '../data/'
cleaned_file = DATA_PATH + 'lending_club_loans_cleaned.csv'

# Leer el CSV limpio (puede tardar un poco)
print("Cargando CSV limpio...")
df_clean = pd.read_csv(cleaned_file, low_memory=False)
print(f"Datos cargados: {df_clean.shape}")

# Configurar conexión a PostgreSQL
# FORMATO: 'postgresql://usuario:contraseña@localhost:5432/nombre_base'
db_user = 'postgres'
db_password = 'admin123'  # <-- CAMBIA ESTO
db_host = 'localhost'
db_port = '5432'
db_name = 'lending_club_db'

engine = create_engine(f'postgresql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}')

# Cargar DataFrame a PostgreSQL (chunksize para no saturar memoria)
print("Subiendo datos a PostgreSQL (esto tomará varios minutos)...")
df_clean.to_sql('loans', engine, if_exists='replace', index=False, chunksize=10000)
print("¡Carga completada exitosamente!")
```
