```python
# Define the path to your data file
DATA_PATH = '../data/' # This goes up one level from the 'notebooks' folder
FILE_NAME = 'accepted_2007_to_2018Q4.csv'  # <-- VERIFICA ESTE NOMBRE

# Load the dataset
try:
    df = pd.read_csv(DATA_PATH + FILE_NAME, low_memory=False)
    print(f"Dataset loaded successfully.")
except FileNotFoundError:
    print(f"Error: File not found at '{DATA_PATH + FILE_NAME}'. Please check the path and file name.")

# Initial Data Exploration
print(f"Initial Shape: {df.shape[0]} rows, {df.shape[1]} columns")
df.head()
```

```python
# Create a copy to avoid modifying the original DataFrame directly
df_cleaned = df.copy()

# List of loan statuses considered as 'completed' with a known outcome
completed_statuses = ['Fully Paid', 'Charged Off', 'Default']

# Filter the DataFrame
df_cleaned = df_cleaned[df_cleaned['loan_status'].isin(completed_statuses)]

print(f"Shape after filtering for completed loans: {df_cleaned.shape[0]} rows")
print(f"Percentage of data retained: {df_cleaned.shape[0] / df.shape[0] * 100:.2f}%")
```

```python
# Identify columns with more than 95% missing values
high_missing_cols = df_cleaned.columns[df_cleaned.isnull().mean() > 0.95].tolist()
print(f"Dropping {len(high_missing_cols)} columns with >95% missing values: {high_missing_cols}")

# Identify columns that are IDs or URLs (often irrelevant for modeling)
id_url_cols = ['id', 'member_id', 'url', 'desc', 'emp_title']
cols_to_drop = high_missing_cols + id_url_cols

# Drop them from our DataFrame
df_cleaned = df_cleaned.drop(columns=[col for col in cols_to_drop if col in df_cleaned.columns])
print(f"Shape after dropping irrelevant columns: {df_cleaned.shape}")
```

```python
# Set a threshold (e.g., 20%)
missing_threshold = 0.20

# Find columns with missing values below the threshold
cols_to_impute = df_cleaned.columns[(df_cleaned.isnull().mean() < missing_threshold) & (df_cleaned.isnull().mean() > 0)].tolist()
cols_to_drop_high_missing = df_cleaned.columns[(df_cleaned.isnull().mean() >= missing_threshold)].tolist()

print(f"Columns to impute (missing < 20%): {len(cols_to_impute)}")
print(f"Columns to drop (missing >= 20%): {len(cols_to_drop_high_missing)}")

df_cleaned = df_cleaned.drop(columns=cols_to_drop_high_missing)
print(f"Shape after applying 80/20 rule: {df_cleaned.shape}")
```

```python
# 1. Separate columns by data type
numeric_cols = df_cleaned.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df_cleaned.select_dtypes(include=['object']).columns.tolist()

print(f"Numerical columns to impute: {len(numeric_cols)}")
print(f"Categorical columns to impute: {len(categorical_cols)}")

# 2. Impute Categorical Columns with 'Unknown'
for col in categorical_cols:
    if df_cleaned[col].isnull().any():
        df_cleaned[col] = df_cleaned[col].fillna('Unknown')

# 3. Impute Numerical Columns with KNNImputer (using a sample for speed)
# KNN can be slow on massive datasets, so we'll use a sample to fit the imputer.
sample_size = 50000 # Adjust based on your PC's performance
if len(df_cleaned) > sample_size:
    df_sample = df_cleaned[numeric_cols].sample(n=sample_size, random_state=42)
else:
    df_sample = df_cleaned[numeric_cols].copy()

print("Fitting KNNImputer on a sample...")
imputer = KNNImputer(n_neighbors=5)
imputer.fit(df_sample)

print("Applying KNNImputer to the entire dataset (this may take a few minutes)...")
# Convert to numpy array for faster transformation
df_cleaned[numeric_cols] = imputer.transform(df_cleaned[numeric_cols])

print("Imputation complete.")
```

```python
# Function to cap outliers using the IQR method
def cap_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
    return df

# Apply capping to key financial features
cols_to_cap = ['loan_amnt', 'annual_inc', 'dti']
for col in cols_to_cap:
    if col in df_cleaned.columns:
        df_cleaned = cap_outliers_iqr(df_cleaned, col)
        print(f"Outliers capped for column: {col}")

# Additional Data Validation: Fix logically impossible values
if 'dti' in df_cleaned.columns:
    df_cleaned['dti'] = df_cleaned['dti'].clip(lower=0, upper=100)
if 'annual_inc' in df_cleaned.columns:
    df_cleaned['annual_inc'] = df_cleaned['annual_inc'].clip(lower=0)
```

```python
# Save the cleaned DataFrame to a new CSV file
cleaned_file_path = DATA_PATH + 'lending_club_loans_cleaned.csv'
df_cleaned.to_csv(cleaned_file_path, index=False)
print(f"Master cleaned dataset saved successfully to '{cleaned_file_path}'")
print(f"Final Shape: {df_cleaned.shape[0]} rows, {df_cleaned.shape[1]} columns")
```
